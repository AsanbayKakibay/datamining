```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Дата': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-03'],
    'Продукт': ['Плов', 'Мантy', 'Самса', 'Плов'],
    'Количество': [10, 15, 5, 8],
    'Цена': [2000, 1500, 500, 2000],
    'Выручка': [20000, 22500, 2500, 16000],
    'Город': ['Алматы', 'Астана', 'Алматы', 'Шымкент']
})

# Описательные статистики
print(data.describe())

# Проверка на пропущенные данные
print(data.isnull().sum())

# Распределение выручки по продуктам (гистограмма)
data.groupby('Продукт')['Выручка'].sum().plot(kind='bar', title='Выручка по продуктам')
plt.show()

# Распределение продаж по городам (круговая диаграмма)
data.groupby('Город')['Количество'].sum().plot(kind='pie', title='Продажи по городам', autopct='%1.1f%%')
plt.show()

# Общее количество проданных единиц продукции
total_units_sold = data['Количество'].sum()
print("Общее количество проданных единиц продукции: " + str(total_units_sold))

# Средняя выручка на заказ
mean_revenue_per_order = data['Выручка'].mean()
print("Средняя выручка на заказ: " + str(mean_revenue_per_order))

# Корреляция между ценой и количеством продаж
correlation_price_quantity = data['Цена'].corr(data['Количество'])
print("Корреляция между ценой и количеством продаж: " + str(correlation_price_quantity))

# Средняя выручка по городам
mean_revenue_per_city = data.groupby('Город')['Выручка'].mean()
print("Средняя выручка по городам:\n" + mean_revenue_per_city.to_string())

# Топ-продукты по выручке
top_products_by_revenue = data.groupby('Продукт')['Выручка'].sum().sort_values(ascending=False)
print("Топ-продукты по выручке:\n" + top_products_by_revenue.to_string())
```
